## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

import matplotlib.pyplot as plt

def random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    return bottom + (ceiling-bottom)*x

uniform_dist1 = random_numbers(10, 15, 100)
plt.hist(uniform_dist1, bins=10)
plt.title("Uniform Distribution (bottom=10, ceiling=15, count=100)")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

```python
dist2 = random_numbers(10, 60, 1000)

plt.hist(dist2, bins=10)
plt.title("Uniform Distribution (bottom=10, ceiling=60, count=1000)")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

How are the two distributions different?

```python
# your answer below

# La segunda distribucion probablemente sea más variada y más amplia, ya que el rango es mucho más amplio (10-60 frente a 10-15)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here

import numpy as np

def normal_random_numbers(mean, std_dev, count):
    return np.random.normal(mean, std_dev, count)

# Le asigno los valores que me piden en el enunciado y la nombro data1

data1 = normal_random_numbers(10, 1, 1000)

# Ahora lo mismo para data2

data2 = normal_random_numbers(10, 50, 1000)

# Ploteo el gráfico
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].hist(data1, bins=30, alpha=0.7, color='blue')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')
axs[0].set_title(f'Normal Distribution with mean=10, std_dev=1')

axs[1].hist(data2, bins=30, alpha=0.7, color='red')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')
axs[1].set_title(f'Normal Distribution with mean=10, std_dev=50')

plt.show()
```

How are the two distributions different?

```python
# your answer below

# La diferencia es que en la primera gráfica los datos están más concentrados en torno a la media, mientras que en la segunda están mucho más dispersos.
# Esto tiene que ver con la desviación típica, cuanto más alta, mayor es la dispersión
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here

def exponential(mean, size):
    return np.random.exponential(scale=mean, size=size)

data3 = exponential(mean=10, size=10)
data4 = exponential(mean=10, size=100)

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(data3, bins=100)
ax1.set_title("Exponential Distribution (n=10)")

ax2.hist(data4, bins=100)
ax2.set_title("Exponential Distribution (n=100)")

plt.show()
```

How are the two distributions different?

```python
# your answer here

# Parece que la segunda gráfica nos da más información, ya que la muestra es más grande.
# Además, la forma de la segunda gráfica se acerca más a la forma teórica de la distribución exponencial.
```
