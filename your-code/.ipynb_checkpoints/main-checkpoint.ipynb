## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
import numpy as np
import pylab as plt
import seaborn as sns
from scipy.stats import expon
from scipy.stats import norm
from scipy.stats import t
from scipy.stats import uniform
```

```python
# your code here
def generate_uniform_random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms
```

```python
# 1. Genero números aleatorios uniformemente distribuidos para el primer conjunto de parámetros
randoms1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)
```

```python
# 2. Genero números aleatorios uniformemente distribuidos para el segundo conjunto de parámetros
randoms2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)
```

```python
# 3. Plotting the histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(randoms1, bins=10)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=15, count=100)')
```

```python
# 3 Plotting the histograms
plt.subplot(1, 2, 2)
plt.hist(randoms2, bins=10)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=60, count=1000)')
```

How are the two distributions different?

```python
# your answer below
"""1) Por empezar estoy utilizando diferentes parametros, entonces esto genera que el "shape" del grafico sea distinto. 2) el "count of values" es distinto, es decir en mi primer histograma se ven reflelados 100 valores en cambio en la segunda 1000. Mi segunda grafica es mas ancha, a pesar que a golpe de vista parece lo contrario, ya que en este caso mi eje x esta representado por muchos mas valores que en mi primer grafico."""
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# 1. your code here
def generate_normal_random_numbers(mean, std_dev, count):
    randoms = np.random.normal(loc=mean, scale=std_dev, size=count)
    return randoms
```

```python
# 2.
randoms1 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)
```

```python
# 2.
randoms2 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)
```

```python
#3
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(randoms1, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution (mean=10, std_dev=1)')
```

```python
#3
plt.subplot(1, 2, 2)
plt.hist(randoms2, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution (mean=10, std_dev=50)')
```

How are the two distributions different?

```python
# your answer below

""" partimos de graficos construidos a partir de una media similar pero diferente desviacion estandar. En la primera distribución con desviación estándar 1, los valores estarán muy agrupados alrededor de la media, lo que dará como resultado una distribución estrecha y alta. En la segunda distribución con una desviación estándar de 50, los valores estarán más dispersos, lo que dará como resultado una distribución más amplia y plana(faltten). """
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def generate_exponential(mean, size):
    return np.random.exponential(scale=mean, size=size)
```

```python
# Generate two sequences with sizes 10 and 100
sequence1 = generate_exponential(mean=10, size=10)
sequence2 = generate_exponential(mean=10, size=100)
```

```python
# Plot the distributions as histograms
plt.hist(sequence1, bins=100, alpha=0.5, label="Size 10")
# Add labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Exponential Distribution")
plt.legend()
```

```python
plt.hist(sequence2, bins=100, alpha=0.5, label="Size 100")

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Exponential Distribution")
plt.legend()
```

How are the two distributions different?

```python
# your answer here
"""Sample Size, en la primera es 10 en la segunda es 100. Spread or Variability, en la segunda tengo los valores mas dispersos. Smoother Distribution, la segunda grafica es mas suave osea menos irregular."""
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)
mean = 10  # Media de la distribución exponencial
rate = 1 / mean  # Tasa de la distribución exponencial

# Calcular la probabilidad P(x < 15)
probability = expon.cdf(15, scale=1/rate)
probability
```

What is the probability that the customer will spend more than 15 minutes

```python
# your answer here
probabilityMore = 1 - expon.cdf(15, scale=1/rate)
probabilityMore
```
